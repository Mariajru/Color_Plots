# ColorChecker patches plots

```python
# Import packages
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import matplotlib.patches as mpatches
from matplotlib.pyplot import figure
import random
```

## Bar plot
Each bar has the color of its corresponding patch in the ColorChecker of 24 patches.

```python
labels = ['Dark skin', 'Light skin', 'Blue sky', 'Foliage', 'Blue flower', 'Bluish green', 'Orange',
          'Purplish blue', 'Moderate red', 'Purple', 'Yellow green', 'Orange yellow',
          'Blue', 'Green', 'Red','Yellow', 'Magenta', 'Cyan', 'White', 'Neutral 8', 'Neutral 6.5', 'Neutral 5',
          'Neutral 3.5', 'Black']

col_array = ('sienna', 'peachpuff', 'steelblue', 'olivedrab', 'mediumpurple', 'aquamarine', 
             'orange', 'royalblue', 'lightcoral', 'rebeccapurple', 'greenyellow', 'gold',
             'darkblue', 'limegreen', 'red', 'yellow', 'magenta', 'dodgerblue', 'white',
             'lightgrey', 'darkgrey', 'grey', 'dimgrey', 'black')

# The width of the bars: can also be len(x) sequence
width = 0.8

# Data 
T_means = [random.randrange(1, 14) for i in range(24)]
Pre_T_means = [random.randrange(1, 14) for i in range(24)]

# PLOT DESIGN
fig, ax = plt.subplots(1, figsize=(14, 8), dpi = 120)
ax = plt.subplot()
# Background color
ax.set_facecolor('whitesmoke')
# Grid
ax.grid(visible=True)
ax.grid(color='lightgrey', linestyle=(0,(5,5)), linewidth=0.4, alpha=0.5)
# Hide the right and top spines
ax.spines.right.set_visible(False)
ax.spines.top.set_visible(False)
ax.spines.bottom.set_visible(False)
ax.spines.left.set_visible(False)
ax.tick_params(which='both', width=0)
# Labels and legend
ax.set_ylabel('$_{00}$', fontsize=20)
ax.bar(labels, T_means, width, label='After Transform', alpha=0.7, color=col_array)
ax.bar(labels, Pre_T_means, width, bottom=T_means,
       label='Before Transform', color='lightsteelblue', alpha=0.7)
ax.set_title('Color differences before and after color calibration', fontsize=15)
patch_1 = mpatches.Patch(color='lightsteelblue', label='Before Transform')
ax.legend(fontsize = 15, handles=[patch_1])
plt.xticks(rotation=65, ha="right", fontsize=15)
plt.yticks(fontsize=15)
plt.ylim([0, 28])

plt.show()
```

## Line plot

```python
col_array = ('sienna', 'peachpuff', 'steelblue', 'olivedrab', 'mediumpurple', 'aquamarine', 
             'orange', 'royalblue', 'lightcoral', 'rebeccapurple', 'greenyellow', 'gold',
             'darkblue', 'limegreen', 'red', 'yellow', 'magenta', 'dodgerblue', 'white',
             'lightgrey', 'darkgrey', 'grey', 'dimgrey', 'black')

Observer_1 = [random.randrange(15, 28) for i in range(24)]
Observer_2 = [random.randrange(5, 16) for i in range(24)]
Observer_3 = [random.randrange(5, 16) for i in range(24)]
Observer_4 = [random.randrange(8, 22) for i in range(24)]

fig, ax = plt.subplots(1, figsize=(14, 8), dpi = 120)
ax.plot(labels, Observer_1, alpha=0.7, marker='o',  color='lightcoral')
ax.plot(labels, Observer_2, alpha=0.7, marker='o',  color='orange')
ax.plot(labels, Observer_3, alpha=0.7, marker='o',  color='olivedrab')
ax.plot(labels, Observer_4, label='Before Transform', color='black', alpha=0.7, marker='o')

# PLOT STYLE
ax = plt.subplot()
# Background color
ax.set_facecolor('whitesmoke')
# Grid
ax.grid(visible=True)
ax.grid(color='lightgrey', linestyle=(0,(5,5)), linewidth=0.4, alpha=0.5)
# Hide the right and top spines
ax.spines.right.set_visible(False)
ax.spines.top.set_visible(False)
ax.spines.bottom.set_visible(False)
ax.spines.left.set_visible(False)
ax.tick_params(which='both', width=0)
# Only show ticks on the left and bottom spines
ax.set_ylabel('$_{00}$', fontsize=20)
ax.set_title('Color differences per observer', fontsize=15)
ax.legend(fontsize = 15)
plt.xticks(rotation=65, ha="right", fontsize=15)
plt.yticks(fontsize=15)
plt.ylim([0, 30])

plt.show()
```
